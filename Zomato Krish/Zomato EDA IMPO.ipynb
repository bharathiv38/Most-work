# Here we can we data has country code, we can merge it by using pd.merge()

```python
final_data = pd.merge(data,data_country,on='Country Code',how='left')
```

```python
country_val=final_data.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_val,labels=country_names)
```

### But here we can see the data is not clearly distributed so lets see top 3 countries

```python
#top 3 countries used by zomato
plt.pie(country_val[:3],labels=country_names[:3])
```

### We need to see the percentage as well so lets us percentage for this we will use autopct

```python
#top 3 countries used by zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

### Observation:

1. Zomato maximum records/transaction are from India.
2. USA & UK has second & third Place

```python
final_data.columns
```

```python
final_data.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
#.size shows the values 
final_data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#reset_index shows the total values of all the rating and we need to rename it here we have given name rating count
final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# we are saving in one variable
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observations

1. When rating is between 4.5 to 4.9 is excellent
2. When rating is between 4 to 4.4 is Good
3. When rating is between 3.5 to 3.9 is Good
4. When rating is between 2.5 to 3.4 is Average
5. When rating is between 1.9 to 2.4 is Poor
6. Max no of peoples have not given ratings

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.set_theme()
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

### Observation

1. Not Rated are count is very high
2. Maximum no of rating are between 2.5 to 3.4

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
